### Feature Selection

#### Remove Constant columns
- Some features always contain the same value.
- These features are useless.

```python
# find constand columns
constant_columns = []

for col in df.columns:
    if df[col].nunique() == 1:
        constant_columns.append(col)

print("Constant Columns: ", constant_columns)
# Constant Columns:  ['Bwd_PSH_Flags', 'Fwd_URG_Flags', 'Bwd_URG_Flags', 'FIN_Flag_Count', 'PSH_Flag_Count', 'ECE_Flag_Count',
#  'Fwd_Avg_Bytes/Bulk', 'Fwd_Avg_Packets/Bulk', 'Fwd_Avg_Bulk_Rate', 'Bwd_Avg_Bytes/Bulk', 'Bwd_Avg_Packets/Bulk', 'Bwd_Avg_Bulk_Rate']
```

```python
print("Before columns drop : ")
print("Train : " ,  df.shape )
df.drop(columns=constant_columns, inplace=True)
print("After columns drop: ")
print("Train : " ,  df.shape )
```

#### Remove Irrelevant Feature

- Flow ID, Source IP, Source Port, Destination IP, Destination Port: These features specify who was talking. If we train a model on these, the model will just memorize the IP addresses of the attackers in the CIC dataset rather than learning how the attack looks. It will fail on a new network.
- Timestamp: Decision trees cannot handle raw datetime strings easily, and an attack happens independently of the exact millisecond recorded in 2018.
- SimillarHTTP: Contains highly unstructured text or is mostly missing.
- 'Fwd_Header_Length.1' and 'Fwd_Header_Length' duplicate feature

```python
columns_to_drop = ['Flow_ID', 'Source_IP', 'Source_Port', 'Destination_IP', 'Destination_Port', 'Timestamp', 'SimillarHTTP', 'Fwd_Header_Length.1']

print("Before columns drop : ")
print("Train : " ,  df.shape )
# drop columns if available
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
print("After columns drop: ")
print("Train : " ,  df.shape )
```

### Binary Label generation

```python
all_labels = ['MSSQL', 'NetBIOS', 'SNMP', 'BENIGN', 'NTP', 'UDP', 
              'Portmap', 'DNS', 'LDAP', 'UDPLag', 'SSDP', 'TFTP']

sampled_frames = []

for label in all_labels:
    subset = df[df['Label'] == label]
    
    if label == 'BENIGN':
        sampled_frames.append(subset)
        print(f"Collected all {len(subset)} Benign rows.")
    else:
        n_to_sample = min(10000, len(subset))
        subset_sampled = subset.sample(n=n_to_sample, random_state=42)
        sampled_frames.append(subset_sampled)
        print(f"Sampled {n_to_sample} rows from {label}.")

df_balanced = pd.concat(sampled_frames)
```

```python

df_balanced['Label'] = np.where(df_balanced['Label'] == 'BENIGN', 0, 1)

df_final = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

df_final.to_csv('../dataset/binary_modal_dataset.csv', index=False)

print(f"\nFinal dataset saved! Total rows: {df_final.shape[0]}")
print(df_final['Label'].value_counts())
```
